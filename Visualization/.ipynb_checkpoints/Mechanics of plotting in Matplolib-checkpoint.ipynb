```python
# Create a Figure object
fig = plt.figure()
```

For creating a new Figure object we can use `plt.figure()` that accepts several parameters. We see that created Figure object is of size 432x288 that represent width and height in inches (6.4x4.8 by default). We can change the size of Figure using `figsize` parameter where we define desired size in the form a tuple.

```python
fig = plt.figure(figsize = (20,15))
```

The size of the Figure object has changed to 1440x1080. Notice that our Figure does not contain any Axes so far. To generate a single Axes object we use `plt.subplots()` when object will be occupy 1 row and 1 column.

```python
# Create a single Figure object and Axes object (subplot)
fig, ax = plt.subplots()
type(ax)
```

We have created a Figure object within wich there is a Axes object labelled as AxesSubplot. Now when we already created Axes you can better see the change of Figure size. 

```python
# Customize the Figure object size
fig, ax = plt.subplots(figsize = (10,8))
```

To generate a set of multiple subplots we define the particular number of rows and columns. For example, if we would like to have 6 subplots while 3 of them should be in one row we would write it like below:

```python
# Create 6 subplots
fig, ax = plt.subplots(2,3)
```

As our subplots are now overlaping but we can easily adjust spacing between the Figure edge and subplots´s edges using `plt.tight_layout()`.

```python
fig, ax = plt.subplots(2,3)
plt.tight_layout()
```

`plt.tight_layout()` can take parameters such `h_pad` and `w_pad` to customize spacing between edges of adjacent subplots.

```python
fig, ax = plt.subplots(2,3)
plt.tight_layout(h_pad = 0.7, w_pad = 3)
```

Now it´s time to create a simple line plot using a generic `.plot()` function. 

```python
import numpy as np
```

```python
# You can run numpy´s random seed to have the same numbers genereted as in this notebook or experiment further
np.random.seed(42)
# Create a sequence of 15 numbers
x = list(range(15))
```

```python
# Examine the output
print(x)
```

```python
# Plot x variable
plt.plot(x)
```

`.plot()` function plots each value of x variable and then draw a straight line between a set of points. 

Let´s plot x variable against y variable and create our first line using Matplotlib (`Line2D`).
First, we define Figure and Axes object and set the size. Then you can call `.plot()` function right onto Axes object to generate a line plot where we specify particular properties to our line such as labels, annotations, specified size, etc.  

___
We want our line to be light green 'dashdotted' line containing red point markers with white color in the middle pointing to each of data points. Phew! That´s a lot of conditions. Indeed, with Matplotlib you can customize every object within the Axes.
Also we would like to give some name to our graph and have labelled Axis to make the graph more readable and understandable for the audience. You might want to set additional vertical or horizontal line to depict some value or range of values, which can be easily done. 

But keep in mind, there is beauty in simplicity.

```python
# Set the random seed to generate exactly the same output as in this notebook
# Generate a sequence of random numbers
np.random.seed(42)
y = np.random.randint(low = 5, high = 50, size = 15)
# Examine the output
print(y)
```

```python
# Create a Figure and Axes objects with defined size
fig, ax = plt.subplots(figsize = (12,9))   

# Plot x against y using plt.plot() function
ax.plot(x,y, color = 'green', linestyle = '-.', linewidth = 2,    # Plot x vs. y, set a color, style and wodth of a line
        alpha = 0.6, marker = '.', markeredgewidth = 7,           # Set a transparency of colors, marker type and its width
        markeredgecolor='red', markerfacecolor = 'white')         # Set a color of a marker and white color within a marker 
                                                                  
ax.set_title('The sequence of numbers', fontsize = 16);           # Set a title of graph and a fontsize
ax.set_xlabel('The values of x', fontsize = 12)                   # Set x-axis label and a fontsize
ax.set_ylabel('The values of y', fontsize = 12)                   # Set y-axis label and a fontsize

ax.axhline(y = 40, linewidth = 2, alpha = 0.5, color = 'black');  # Set horizontal line emphasises a value of y-xis
plt.xticks(np.arange(start = 0, stop = len(x), step = 1))
plt.yticks(np.arange(start = 0, stop = max(y), step = 2));
```

The frequency of ticks are generated by Matplotlib in the way that is the most suitable given the data. We can see that ticks on x-axis start at 0, end at 14 while the values are increasing by 2. The y-axis is from 15 till 45 by 5 steps. If we would like to customize it we can use `plt.xticks` and `plt.yticks` to do so. Take a closer look at the original ticks and then uncomment those two rows to see the difference. The new positions of ticks are set using `numpy´s` `arange()` function where we define start and end of interval, and desired spacing between values. In case of x-axis, tick´s stopping point is defined as the length of the variable x and values increase set by one. The last tick on the y-axis is defined as maximum value of the variable y plus 2 aditional numbers. Try to change these values for yourself to see how the position of ticks will adjust.

```python
x_var = np.random.normal(size = 150)
```

To create a histogram of variable we can use `hist()` function where a set of parameters can be specified. See the [documentation](https://matplotlib.org/3.3.3/api/_as_gen/matplotlib.pyplot.hist.html). 

```python
fig, ax = plt.subplots(1,2, figsize = (13,9))
ax[0].hist(x_var, bins = 15, color = 'skyblue')
ax[0].set_xlabel('Values')
ax[0].set_ylabel('Count')
ax[0].set_title('The distribution of data points')

ax[1].hist(x_var, bins = 10, density = True)
ax[1].set_xlabel('Values')
ax[1].set_ylabel('Density')
ax[1].set_title('The density of data points')

plt.tight_layout(pad = 6)
fig.suptitle('Histograms of x_var', x = 0.5, y = 0.97, fontsize = 20);
```

**Description**

Where the default number of bins is 10.
